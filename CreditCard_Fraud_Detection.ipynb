Import the Dependecies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Loading the data


```python
credit_card_data = pd.read_csv('/content/creditcard.csv')
```

```python
# first 5 rows of data set
credit_card_data.head()
```

```python
credit_card_data.tail()
```

```python
# dataset information
credit_card_data.info()
```

```python
# checking the number of missing value in each column

credit_card_data.isnull().sum()
```

```python
credit_card_data.shape
```

```python
credit_card_data.isnull().sum().sum()
```

```python
# handling the null values
credit = credit_card_data.fillna(value = 0)
```

```python
credit.isnull().sum()
```

```python
# distribution of legit transaction and fraud transaction
credit['Class'].value_counts()
```

This dataset is highly unbalanced
0 --> Legit Transactions
1 --> Fraud Transaction


```python
# seperating the data for analysis
legit = credit[credit.Class == 0]
fraud = credit[credit.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
# statistical measures of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# compare the values for both transactions
credit.groupby('Class').mean()
```

Under-sampling


Build a sample dataset containig similar distribution of normal transaction and fraud transaction

Number of fraud transaction -->213

```python
legit_sample = legit.sample(n = 213)
```

axis = 0 means rows
axis = 1 means column

```python
# concatanating two data frames
new_dataset = pd.concat([legit_sample, fraud],axis = 0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

Splitting the data into features & targets

```python
X = new_dataset.drop(columns='Class',axis = 1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

Split the training data and testing data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size = 0.2,stratify= Y,random_state = 2 )
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

Model Training
Logistic Regression

```python
model = LogisticRegression()
```

```python
# trainig the logistic regression model with training data
model.fit(X_train, Y_train)
```

```python
# model Evaulation
# Accuracy score
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction,Y_train)
```

```python
print('Accuracy on trainig data :', training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction,Y_test)
```

```python
print('Accuracy on test data :', test_data_accuracy)
```
